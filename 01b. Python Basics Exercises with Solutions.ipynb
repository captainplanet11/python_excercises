##### 8. What's the output of the following code? 
````

````
#### `PLEASE DON'T CODE`

`````scala
t1 = (1, 2, 4, 3)
t2 = (1, 2, 3, 4)
t1 < t2

# When doing GT, GE, LT, LE the comparison will element-wise. For example, in this case:
# Step 1: Comparison between first elements: 1 < 1 - inconclusive, no one is winnning here
# Step 2: Comparison between second elements: 2 < 2 - inconclusive, no one is winnning here
# Step 3: Comparison between third elements: 4 < 3 - Conclusive and False, it will stop here, it will not do the step 4

# other example: 
#t1 = (1, 1, 5, 5)
#t2 = (1, 2, 3, 4)
# t1 < t2 it will return True because in the second element 1 < 2. Doesn't matter the others elements (3 and 4)

# When we use "==" or "!=" it's a little bit different because to return "True" all the comparisons must be "True"
`````
