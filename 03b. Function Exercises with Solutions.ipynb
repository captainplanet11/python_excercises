`````scala
## 1)


def convert_logic(list_l):
    return list(map(lambda x: x == False, list_l))

def convert_float(list_s):
    return list(map(lambda x: float(x), list_s))
    
def CheckString(data): 
    return [isinstance(x, str) for x in data]
    
    
def Get_Floats(data): 
    mask = convert_logic(CheckString(data))
    values = convert_float(np.array(data)[mask])
    return values
 
Get_Floats(data)  
`````

a) Using list comprehension and the `random` method, generate a list of 100 random (continuous) probability values (e.g. \\(x \in [0,1)\\)).

b) Using functional programming, create a function to convert a list of continuous probabilities into a list of discrete (binary) using some threshold (e.g. probabilities higher than \\(0.5\\) are labeled \\(1\\)). 


`````scala


import random

def RandomProbs(length): 
    '''
    Generates a list of random probabilty values
    '''
    return [round(random.random(),2) for i in range(length)]
    
def ConvertToBinary(y,thresh=0.5): 
    '''
    Converts a list of continuous probs into discrete (binary) values
    '''
    return list(map(lambda x: 1 if x > thresh else 0, y))

probs = RandomProbs(100)
    
print('continuous probs: ', probs)
print(' ')
print('discrete probs: ', ConvertToBinary(probs,.5))
`````

Create a function to compute the accuracy score and compare its results with sklearn's built-in method.

<b>Note</b>: consider the two following lists of values: 

\\(\mathrm{\hat{y}} = [1,2,3,3,2,1,1,2,1]\\)
\\(\mathrm{y} = [1,2,3,4,3,2,1,2,1]\\)



`````scala


y1 = [1,2,3,3,2,1,1,2,1]
y2 = [1,2,3,4,3,2,1,2,1]

def ComputeAccuracy(y_true,y_pred): 
    '''
    Alternative approach to compute the accuracy of a model
    '''
    f = lambda x: 1 if x[0] == x[1] else 0
    x = zip(y_true,y_pred)
    trues = list(map(f,x))
    return float(sum(trues))/len(trues)

from sklearn.metrics import accuracy_score
print(accuracy_score(y1,y2), ComputeAccuracy(y1,y2))
`````
